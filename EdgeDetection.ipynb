```python
img = cv2.imread('Cat.bmp')
img2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

plt.figure(figsize=figsize)
plt.imshow(img2,cmap="gray",vmin=0,vmax=255)
plt.title("Grayscale Image")
plt.show()
```

```python
# Using Prewitt mask in X axis

m= img2.shape[0]
n= img2.shape[1]
#print(m)
#print(n)

mask = np.array([[-1,-1,-1],[0,0,0,],[1,1,1]])
img_newXP = np.zeros([m,n])
#Convolve the 3x3 Mask Over The Image
img_newXP = np.zeros([m, n])
for i in range(1, m-1):
  for j in range(1, n-1):
      temp = img2[i-1,j-1]*mask[0,0]+ img2[i-1,j]*mask[0,1] + img2[i-1,j+1]*mask[0,2] + img2[i,j-1]*mask[1,0] + img2[i,j]*mask[1,1] + img2[i,j+1]*mask[1,2] + img2[i+1,j-1]*mask[2,0] + img2[i+1,j]*mask[2,1] + img2[i+1,j+1]*mask[2,2]
      img_newXP[i,j]=temp
     

plt.figure(figsize=figsize)
plt.imshow(img_newXP, cmap="gray",vmin=0,vmax=256)
plt.title("Prewitt (X direction)")
plt.show()
```

```python
# Using Prewitt mask in Y axis

m= img2.shape[0]
n= img2.shape[1]
#print(m)
#print(n)

mask = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
img_newYP = np.zeros([m,n])
#Convolve the 3x3 Mask Over The Image
img_newYP = np.zeros([m, n])
for i in range(1, m-1):
  for j in range(1, n-1):
      temp = img2[i-1,j-1]*mask[0,0]+ img2[i-1,j]*mask[0,1] + img2[i-1,j+1]*mask[0,2] + img2[i,j-1]*mask[1,0] + img2[i,j]*mask[1,1] + img2[i,j+1]*mask[1,2] + img2[i+1,j-1]*mask[2,0] + img2[i+1,j]*mask[2,1] + img2[i+1,j+1]*mask[2,2]
      img_newYP[i,j]=temp
      

plt.figure(figsize=figsize)
plt.imshow(img_newYP, cmap="gray",vmin=0,vmax=256)
plt.title("Prewitt (Y direction) ")
plt.show()
```

```python
#Addition of X and Y direction Prewitt

img_newBP=img_newXP+img_newYP
plt.figure(figsize=figsize)
plt.imshow(img_newBP,cmap='gray',vmin=0,vmax=255)
plt.title("Prewitt image in both direction")
plt.show()
```

```python
# Using Sobel mask in X axis
m= img2.shape[0]
n= img2.shape[1]
#print(m)
#print(n)

mask = np.array([[-1,-2,-1],[0,0,0],[1,2,1]])
img_newXS = np.zeros([m,n])
#Convolve the 3x3 Mask Over The Image
img_newXS = np.zeros([m, n])
for i in range(1, m-1):
  for j in range(1, n-1):
      temp = img2[i-1,j-1]*mask[0,0]+ img2[i-1,j]*mask[0,1] + img2[i-1,j+1]*mask[0,2] + img2[i,j-1]*mask[1,0] + img2[i,j]*mask[1,1] + img2[i,j+1]*mask[1,2] + img2[i+1,j-1]*mask[2,0] + img2[i+1,j]*mask[2,1] + img2[i+1,j+1]*mask[2,2]
      img_newXS[i,j]=temp
     

plt.figure(figsize=figsize)
plt.imshow(img_newXS, cmap="gray",vmin=0,vmax=256)
plt.title("Sobel (X direction)")
plt.show()
```

```python
# Using Sobel mask in Y axis
m= img2.shape[0]
n= img2.shape[1]
#print(m)
#print(n)

mask = np.array([[-1,0,1],[-2,0,2],[-1,0,1]])
img_newYS = np.zeros([m,n])
#Convolve the 3x3 Mask Over The Image
img_newYS = np.zeros([m, n])
for i in range(1, m-1):
  for j in range(1, n-1):
      temp = img2[i-1,j-1]*mask[0,0]+ img2[i-1,j]*mask[0,1] + img2[i-1,j+1]*mask[0,2] + img2[i,j-1]*mask[1,0] + img2[i,j]*mask[1,1] + img2[i,j+1]*mask[1,2] + img2[i+1,j-1]*mask[2,0] + img2[i+1,j]*mask[2,1] + img2[i+1,j+1]*mask[2,2]
      img_newYS[i,j]=temp
     

plt.figure(figsize=figsize)
plt.imshow(img_newYS, cmap="gray",vmin=0,vmax=256)
plt.title("Sobel (Y direction)")
plt.show()
```
