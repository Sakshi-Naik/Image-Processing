```python
im = cv2.imread('Peacock.bmp')
plt.figure(figsize = figsize)
plt.imshow(im, cmap='gray', vmin=0,vmax=255)
plt.title("Orginal Image")
plt.show()
```

```python
img = cv2.cvtColor(np.float32(im), cv2.COLOR_BGR2GRAY)
im2 = cv2.resize(img, (256 ,256))

plt.figure(figsize=figsize)
plt.imshow(img,cmap='gray',vmin=0,vmax=256)
plt.title('Grayscale Image')
plt.show()
```

```python
# image negative operation #
[row, cols] = img.shape
neg_img = np.zeros((row,cols))
for r in range(0, row - 1):
  for c in range(0, cols - 1):
    neg_img[r, c] = 255 - img[r, c]  

plt.figure(figsize=figsize)
plt.imshow(img,cmap='gray',vmin=0,vmax=256)
plt.title('Grayscale Image')
plt.show()

plt.figure(figsize=figsize)
plt.imshow(neg_img,cmap='gray',vmin=0,vmax=256)
plt.title('Negative Image')
plt.show()
```

```python
# Log transform Operation #
c=255/np.log(1+np.max(img))
log_img = c*(np.log(img+1))

plt.figure(figsize=figsize)
plt.imshow(img,cmap='gray',vmin=0,vmax=256)
plt.title('Grayscale Image')
plt.show()

plt.figure(figsize=figsize)
plt.imshow(neg_img,cmap='gray',vmin=0,vmax=256)
plt.title('Log transformed Image')
plt.show()
```

```python
# Power law trnsform operation #
gamma = 0.5
g_img = np.array(255*(img/255)**gamma, dtype='uint8')

plt.figure(figsize=figsize)
plt.imshow(img,cmap='gray',vmin=0,vmax=256)
plt.title('Grayscale Image')
plt.show()

plt.figure(figsize=figsize)
plt.imshow(g_img,cmap='gray',vmin=0,vmax=256)
plt.title('Power Law transformed Image')
plt.show()
```

```python
# Bit Plane Slicing #
im3 = cv2.imread('Peacock.bmp', 0)
out = []
row = im3.shape[0]
col = im3.shape[1]
for k in range(0, 7):
  plane = np.full((row, col), 2 ** k, np.uint8)
  res = cv2.bitwise_and(plane, im3)
  x = res *255
  out.append(x)
cv2_imshow(np.hstack(out))
```
