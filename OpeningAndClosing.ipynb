**Practical 8: Opening and Closing Operation on Binary Image**

```python
## Opening and Closing Operation on Binary Image ##
import cv2
import numpy as np 
import sys
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
%matplotlib inline
from io import BytesIO
from PIL import Image
from google.colab import files
from skimage import color
from skimage import io
#uploaded = files.upload()
from PIL import ImageDraw
from PIL import ImageFont
from skimage.util import random_noise
figsize = (6,6)
```

```python
# Creating a binary image for opening operation
img= np.zeros([256,256])
img[30:90,30:90]=1
img[91:150,90:150]=1
img[90,90]=1

img[30:90,151:210]=1
img[90,150]=1

plt.figure(figsize=figsize)
plt.imshow(img, cmap="gray",vmin=0,vmax=1)
plt.title("BINARY IMAGE")
plt.show()
```

```python
##Erosion for Opening Operation

m=img.shape[0]
n=img.shape[1]
struct_ele=np.array([[1,1,1],[1,1,1],[1,1,1]])
img2=np.zeros([m,n])
for i in range(1,m-1):
  for j in range(1,n-1):
     temp = img[i-1,j-1]*struct_ele[0,0], img[i-1,j]*struct_ele[0,1] , img[i-1,j+1]*struct_ele[0,2] , img[i,j-1]*struct_ele[1,0] , img[i,j]*struct_ele[1,1] , img[i,j+1]*struct_ele[1,2] , img[i+1,j-1]*struct_ele[2,0] , img[i+1,j]*struct_ele[2,1] , img[i+1,j+1]*struct_ele[2,2]
     img2[i, j]= min(temp)

plt.figure(figsize=figsize)
plt.imshow(img2,cmap='gray',vmin=0,vmax=1)
plt.title("ERODED IMAGE")
plt.show()
```

```python
## Dilation for Opening operation

m=img.shape[0]
n=img.shape[1]
struct_ele=np.array([[1,1,1],[1,1,1],[1,1,1]])
img3=np.zeros([m,n])
for i in range(1,m-1):
  for j in range(1,n-1):
     temp =[ img2[i-1,j-1]*struct_ele[0,0], img2[i-1,j]*struct_ele[0,1] , img2[i-1,j+1]*struct_ele[0,2] , img2[i,j-1]*struct_ele[1,0] , img2[i,j]*struct_ele[1,1] , img2[i,j+1]*struct_ele[1,2] , img2[i+1,j-1]*struct_ele[2,0] , img2[i+1,j]*struct_ele[2,1] , img2[i+1,j+1]*struct_ele[2,2] ]
     img3[i, j]= max(temp)

plt.figure(figsize=figsize)
plt.imshow(img3,cmap='gray',vmin=0,vmax=1)
plt.title("OPENED IMAGE")
plt.show()
```

```python
## Creating binary image for Closing operation
img4= np.zeros([256,256])
img4[30:90,30:90]=1

img4[89:150,91:150]=1

img4[30:90,151:210]=1

plt.figure(figsize=figsize)
plt.imshow(img4, cmap="gray",vmin=0,vmax=1)
plt.title("BINARY IMAGE 03")
plt.show()
```

```python
## Dilation for Closing operation
m=img.shape[0]
n=img.shape[1]
struct_ele=np.array([[1,1,1],[1,1,1],[1,1,1]])
img5=np.zeros([m,n])
for i in range(1,m-1):
  for j in range(1,n-1):
     temp =[ img4[i-1,j-1]*struct_ele[0,0], img4[i-1,j]*struct_ele[0,1] , img4[i-1,j+1]*struct_ele[0,2] , img4[i,j-1]*struct_ele[1,0] , img4[i,j]*struct_ele[1,1] , img4[i,j+1]*struct_ele[1,2] , img4[i+1,j-1]*struct_ele[2,0] , img4[i+1,j]*struct_ele[2,1] , img4[i+1,j+1]*struct_ele[2,2] ]
     img5[i, j]= max(temp)

plt.figure(figsize=figsize)
plt.imshow(img5,cmap='gray',vmin=0,vmax=1)
plt.title("DILATED IMAGE")
plt.show()
```

```python
## Erosion for closing operation

m=img.shape[0]
n=img.shape[1]
struct_ele=np.array([[1,1,1],[1,1,1],[1,1,1]])
img6=np.zeros([m,n])
for i in range(1,m-1):
  for j in range(1,n-1):
     temp = img5[i-1,j-1]*struct_ele[0,0], img5[i-1,j]*struct_ele[0,1] , img5[i-1,j+1]*struct_ele[0,2] , img5[i,j-1]*struct_ele[1,0] , img5[i,j]*struct_ele[1,1] , img5[i,j+1]*struct_ele[1,2] , img5[i+1,j-1]*struct_ele[2,0] , img5[i+1,j]*struct_ele[2,1] , img5[i+1,j+1]*struct_ele[2,2]
     img6[i, j]= min(temp)

plt.figure(figsize=figsize)
plt.imshow(img6,cmap='gray',vmin=0,vmax=1)
plt.title("CLOSING IMAGE")
plt.show()
```
