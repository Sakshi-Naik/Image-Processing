**Practical 3: Plotting Histogram and Histogram Equalization**

```python
# Plotting Histogram and Histogram Equalization #
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
from io import BytesIO
from PIL import Image
from google.colab import files
from skimage import color
from skimage import io
from PIL import ImageDraw
from skimage.util import random_noise
figsize = (6, 6)
```

```python
uploaded=files.upload()
```

```python
# Reading Image #
im = cv2.imread('Peacock.bmp')
im1=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
img = cv2.resize(im, (256,256))
plt.figure(figsize = figsize)
plt.imshow(img, cmap="gray", vmin=0, vmax=255)
plt.title("Orignal Image")
plt.show()
```

```python
bin_edges_min_max = [0,256]
num_bins = 256
bin_count1, bin_edges1 = np.histogram(img, num_bins, bin_edges_min_max)
bin_start1 = bin_edges1[:-1]

def draw_histo(x_axis, input):
  fig, ax = plt.subplots(figsize = figsize)
  plt.bar(x_axis, input, 256/(x_axis[-1]-x_axis[0]+1))
  return fig, ax
draw_histo(bin_start1, bin_count1)
```

```python
# Plotting Histogram #
I3 = np.zeros(256, np.int32)
for i in range (256):
    for j in range (256):
      r = img[i][j]
      I3[r] = I3[r]+1
plt.title('Histogram plot')
plt.plot(I3)
```

```python
pd = np.zeros(256)
for ii in range (256):
  pd[ii] = I3[ii]/(256*256)
cdf = np.cumsum(pd)
f_eq = np.round(cdf * 255).astype(int)
I_eq = f_eq[img]
```

```python
bin_edges_min_max = [0, 256]
num_bins = 256
bin_count1, bin_edges1 = np.histogram(img, num_bins, bin_edges_min_max)
bin_start1 = bin_edges1[:-1]

draw_histo(bin_start1, bin_count1)
plt.title("Equalized Histogram")
plt.show()

plt.figure(figsize = figsize)
plt.title("Equalized Image")
plt.imshow(I_eq)
```
