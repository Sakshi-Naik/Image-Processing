```python
# Seperating Planes
B=im[:,:,0]
G=im[:,:,1]
R=im[:,:,2]
ax1 = np.concatenate((B,G,R), axis=1)
cv2_imshow(ax1)
r1=im.copy()
g1=im.copy()
b1=im.copy()
r1[:,:,(0,1)]=0
g1[:,:,(0,2)]=0
b1[:,:,(1,2)]=0
ax2 = np.concatenate((r1,g1,b1), axis=1)
cv2_imshow(ax2)
```

```python
# Multiplication Operation
B=im[:,:,0]
G=im[:,:,1]
R=im[:,:,2]
im5 = 0.298*R+0.578*G+0.114*B
cv2_imshow(im5)
```

```python
# Division Operation
B=im[:,:,0]
G=im[:,:,1]
R=im[:,:,2]
im5 = (0.298*R+0.578*G+0.114*B) / 4
cv2_imshow(im5)
```

```python
# Addition Operation
im6=cv2.add(im5,100)
ax3=np.concatenate((im5,im6),axis=1)
cv2_imshow(ax3)
```

```python
# Subtraction Operation
im7=cv2.subtract(im5,100)
ax4=np.concatenate((im5,im7),axis=1)
cv2_imshow(ax4)
```
