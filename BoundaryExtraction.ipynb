**Practical 7:  Boundry Extraction using Morphological Operation on Binary Image**

```python
## Boundry Extraction using Morphological Operation on Binary Image ##
import cv2
import numpy as np 
import sys
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
%matplotlib inline
from io import BytesIO
from PIL import Image
from google.colab import files
from skimage import color
from skimage import io
#uploaded = files.upload()
from PIL import ImageDraw
from PIL import ImageFont
from skimage.util import random_noise
figsize = (6,6)
```

```python
# Creating a binary image
img=np.zeros([256,256])


img[50:100, 20:40]=1
img[50:100, 50:70]=1
img[50:100, 80:100]=1
img[50:100, 110:130]=1

img[150:200, 20:40]=1
img[150:200, 50:70]=1
img[150:200, 80:100]=1
img[150:200, 110:130]=1

plt.figure(figsize=figsize)
plt.imshow(img,cmap="gray",vmin=0,vmax=1)
plt.title('Binary Image')
plt.show()
```

```python
#EROSION OPERATION

m=img.shape[0]
n=img.shape[1]
strE=np.array([[1,1,1],[1,1,1],[1,1,1]])
img1=np.zeros([m,n])
for i in range(1,m-1):
  for j in range(1,n-1):
     temp = img[i-1,j-1]*strE[0,0], img[i-1,j]*strE[0,1] , img[i-1,j+1]*strE[0,2] , img[i,j-1]*strE[1,0] , img[i,j]*strE[1,1] , img[i,j+1]*strE[1,2] , img[i+1,j-1]*strE[2,0] , img[i+1,j]*strE[2,1] , img[i+1,j+1]*strE[2,2]
     img1[i, j]= min(temp)

plt.figure(figsize=figsize)
plt.imshow(img1,cmap='gray',vmin=0,vmax=1)
plt.title("ERODED Image")
plt.show()
```

```python
# BOUNDRY EXTRACTION
img3=img-img1
plt.figure(figsize=figsize)
plt.imshow(img3,cmap='gray',vmin=0,vmax=1)
plt.title("BOUNDARY EXTRACTED IMAGE")
plt.show()
```
