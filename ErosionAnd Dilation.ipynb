```python
## Erosion and Dilation on Binary and Greyscale Image ##
import cv2
import numpy as np 
import sys
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
%matplotlib inline
from io import BytesIO
from PIL import Image
from google.colab import files
from skimage import color
from skimage import io
#uploaded = files.upload()
from PIL import ImageDraw
from PIL import ImageFont
from skimage.util import random_noise
figsize = (6,6)
```

```python
# Creating a binary image
img=np.zeros([256,256])


img[50:100, 20]=1
img[50:100, 50]=1
img[50:100, 70]=1
img[50:100, 90]=1

img[150:200, 20]=1
img[150:200, 50]=1
img[150:200, 70]=1
img[150:200, 90]=1

plt.figure(figsize=figsize)
plt.imshow(img,cmap="gray",vmin=0,vmax=1)
plt.title('Binary Image')
plt.show()
```

```python
# Dilation
m=img.shape[0]
n=img.shape[1]
struct_ele=np.array([[1,1,1],[1,1,1],[1,1,1]])
img2=np.zeros([m,n])
for i in range(1,m-1):
  for j in range(1,n-1):
     temp =[ img[i-1,j-1]*struct_ele[0,0], img[i-1,j]*struct_ele[0,1] , img[i-1,j+1]*struct_ele[0,2] , img[i,j-1]*struct_ele[1,0] , img[i,j]*struct_ele[1,1] , img[i,j+1]*struct_ele[1,2] , img[i+1,j-1]*struct_ele[2,0] , img[i+1,j]*struct_ele[2,1] , img[i+1,j+1]*struct_ele[2,2] ]
     img2[i, j]= max(temp)

plt.figure(figsize=figsize)
plt.imshow(img2,cmap='gray',vmin=0,vmax=1)
plt.title("DILATED IMAGE")
plt.show()
```

```python
# Erosion
m=img.shape[0]
n=img.shape[1]
strE=np.array([[1,1,1],[1,1,1],[1,1,1]])
img3=np.zeros([m,n])
for i in range(1,m-1):
  for j in range(1,n-1):
     temp = img[i-1,j-1]*strE[0,0], img[i-1,j]*strE[0,1] , img[i-1,j+1]*strE[0,2] , img[i,j-1]*strE[1,0] , img[i,j]*strE[1,1] , img[i,j+1]*strE[1,2] , img[i+1,j-1]*strE[2,0] , img[i+1,j]*strE[2,1] , img[i+1,j+1]*strE[2,2]
     img3[i, j]= min(temp)

plt.figure(figsize=figsize)
plt.imshow(img3,cmap='gray',vmin=0,vmax=1)
plt.title("ERODED IMAGE")
plt.show()
```

```python
## Greyscale Image ##
im2=cv2.imread('Cat.bmp')
im3=cv2.cvtColor(im2,cv2.COLOR_BGR2GRAY)

plt.figure(figsize=figsize)
plt.imshow(im3, cmap="gray",vmin=0,vmax=255)
plt.title("GRAYSCALE IMAGE")
plt.show()
```

```python
#DILATION OPERATION ON GRAYSCALE IMAGE

m=img.shape[0]
n=img.shape[1]
strE=np.array([[1,1,1],[1,1,1],[1,1,1]])
im4=np.zeros([m,n])
for i in range(1,m-1):
  for j in range(1,n-1):
     temp =[ im3[i-1,j-1]*strE[0,0], im3[i-1,j]*strE[0,1] , im3[i-1,j+1]*strE[0,2] , im3[i,j-1]*strE[1,0] , im3[i,j]*strE[1,1] , im3[i,j+1]*strE[1,2] , im3[i+1,j-1]*strE[2,0] , im3[i+1,j]*strE[2,1] , im3[i+1,j+1]*strE[2,2] ]
     im4[i, j]= max(temp)

plt.figure(figsize=figsize)
plt.imshow(im4,cmap='gray',vmin=0,vmax=255)
plt.title("DILATED GRAYSCALE IMAGE")
plt.show()
```
